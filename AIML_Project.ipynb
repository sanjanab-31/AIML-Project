```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler

df = pd.read_excel("assets/personalized_training_dataset.xlsx")

categorical_cols = ["Gender", "Sport", "Fitness_Level", "Primary_Goal", "Injury_History", "Preferred_Training_Type", "Improvement_Area"]
encoder = LabelEncoder()
for col in categorical_cols:
    df[col] = encoder.fit_transform(df[col])

numeric_cols = ["Age", "Height_cm", "Weight_kg", "Weekly_Training_Hours", "Past_Training_Load", "Rest_Days_Per_Week", "Performance_Score"]
scaler = StandardScaler()
df[numeric_cols] = scaler.fit_transform(df[numeric_cols])

print("Missing Values:", df.isnull().sum())
```
